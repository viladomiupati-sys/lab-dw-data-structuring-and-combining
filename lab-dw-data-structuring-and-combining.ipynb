## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

# incorporar data frame 1
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

insurance_df = pd.read_csv(url)
insurance_df

# netejant data frame 1
def clean_column_names(df):
    df = df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'))
    df = df.rename(columns={'st': 'state'})
    df = df.rename(columns={'customer': 'customer_id'})
    df = df.rename(columns={'education': 'education_level'})
    df = df.rename(columns={'monthly_premium_auto': 'monthly_auto_premium'})
    df = df.rename(columns={'number_of_open_complaints': 'open_complaints_count'})
    return df

# comprobant que estigui neta
insurance_df = clean_column_names(insurance_df)
insurance_df

# eliminant totes les files que son Nan values de la primera data frame
insurance_df = insurance_df.drop(insurance_df.tail(2937).index)
insurance_df

# incortorant data frame 2
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"

insurance_df_2 = pd.read_csv(url)
insurance_df_2.head()

# cleaning data frame 2
def clean_column_names(df):
    df = df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'))
    df = df.rename(columns={'st': 'state'})
    df = df.rename(columns={'customer': 'customer_id'})
    df = df.rename(columns={'education': 'education_level'})
    df = df.rename(columns={'monthly_premium_auto': 'monthly_auto_premium'})
    df = df.rename(columns={'number_of_open_complaints': 'open_complaints_count'})
    return df

# comprobant que estigui neta
insurance_df_2 = clean_column_names(insurance_df_2)
insurance_df_2

# incorporant 3
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

insurance_df_3 = pd.read_csv(url)
insurance_df_3

# netajant data frame 3
def clean_column_names(df):
    df = df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'))
    df = df.rename(columns={'st': 'state'})
    df = df.rename(columns={'customer': 'customer_id'})
    df = df.rename(columns={'education': 'education_level'})
    df = df.rename(columns={'monthly_premium_auto': 'monthly_auto_premium'})
    df = df.rename(columns={'number_of_open_complaints': 'open_complaints_count'})
    return df

insurance_df_3 = clean_column_names(insurance_df_3)
insurance_df_3

# unint les tres taules
df_total = pd.concat([insurance_df, insurance_df_2, insurance_df_3], ignore_index=True)
df_total


# Cleaning invalid values
# Gender
df_total['gender'].unique()

df_total['gender'] = df_total['gender'].str.strip().str.upper()  # Remove spaces & unify case
df_total['gender'] = df_total['gender'].replace({
    'FEMALE': 'F',
    'FEMAL': 'F',
    'MALE': 'M'
})

# State
df_total['state'].unique()

state_map = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'CALI': 'California',
    'WA': 'Washington',
    'NV':'Nevada',
    'OR':'Oregon'
}

df_total['state'] = df_total['state'].replace(state_map)
df_total['state'] = df_total['state'].str.strip()
df_total['state'].unique()

# Education
df_total['education_level'].unique()

education_map = {
    'Bachelors': 'Bachelor',
}

df_total['education_level'] = df_total['education_level'].replace(education_map)
df_total['education_level'].unique()

# Customer Lifetime value - deliting character %
df_total['customer_lifetime_value'] = (df_total['customer_lifetime_value'].astype(str).str.replace('%', '', regex=False))
df_total['customer_lifetime_value'] = pd.to_numeric(df_total['customer_lifetime_value'], errors='coerce')

# Vehicle Class
df_total['vehicle_class'] = df_total['vehicle_class'].replace({
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
})
df_total['vehicle_class'].unique()

# 3. Formatting data types
# Open complaints counts
#Clearing data
df_total['open_complaints_count'] = df_total['open_complaints_count'].apply(lambda x: x.split('/')[1] if isinstance(x, str) and '/' in x else x)
df_total['open_complaints_count'].unique()

# Chaning to intiger
df_total['open_complaints_count'] = pd.to_numeric(df_total['open_complaints_count'], errors='coerce').astype('Int64')
df_total['open_complaints_count'].dtype

# comprovant com queda la taula
df_total
```

```python
# incorporar la nova data frame
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

marketing_customer = pd.read_csv(url)
marketing_customer
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# fem pivot amb sales i revenue

pivot_revenue = pd.pivot_table(marketing_customer,
    values = 'total_claim_amount',
    index = 'sales_channel',
    aggfunc = 'sum'
).round(2).sort_values(by='total_claim_amount', ascending=False)

pivot_revenue
```

# conclusions:
- The Agents channel records the highest total claim amount (≈ 1.81M), positioning it as the top-performing sales channel in terms of monetary value.
- Branches show solid performance with ≈ 1.30M.
- The Call Center (≈ 0.93M), indicating potential for improvement or targeted optimization strategies.
- The Web channel has the lowest total claim amount (≈ 0.71M), which may reflect lower engagement.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_table2 = pd.pivot_table(marketing_customer,
    index = ['gender', 'education'],
    values ='customer_lifetime_value', # columna per analitzar
    aggfunc = 'mean'                # funció agregada: avg             
).round(2)                           # arodonim a 2 decimals

pivot_table2
```

# conclusions: 
- Highest CLV segments: Women with High School or Below (8,675.22) and Master (8,157.05) or Men with Master (8168.83) or College degrees (8052.46), indicating these groups generate the most value.
- Gender trends: Men show consistently high CLV in Master and College categories, while women exhibit an unexpected peak in High School or Below.
- Insights: Target retention and upselling efforts to Master, College, and high-CLV female segments. Investigate the drivers behind the high CLV in women with lower education for potential replication across other segments.

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
pivot_table3 = pd.pivot_table(marketing_customer,
    index = ['policy_type', 'month'],
    values = 'number_of_open_complaints',
    aggfunc = 'sum'
).reset_index().sort_values(by='number_of_open_complaints', ascending=False)
pivot_table3
```
